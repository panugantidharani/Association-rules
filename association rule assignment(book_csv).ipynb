# Data Mining

```python
#import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#import Dataset
book=pd.read_csv('book.csv')
book
```

```python
#data preprocessing not required as it is already in transaction format
```

# Apriori Algorithm

# 1.Association rule with 10% support and 70% confidence

```python
#with 10% support
frequent_itemsets=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
#with 70% confidence
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
# A levarage value of 0 indicates independence.Range will be [-1,1]
# A high conviction value means that the consequent is highly depending on the data
```

```python
rules.sort_values('lift',ascending=False)
```

```python
#lift ratio>1 is a good influential rule in selecting the associated transaction
rules[rules.lift>1]
```

```python
#visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
